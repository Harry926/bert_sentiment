```python
tweet_id = []

t = client.get_users_tweets(id = USER_ID, max_results = 50, tweet_fields = ['conversation_id', 'in_reply_to_user_id'])
for i in t.data:
    #print(i.id, i.conversation_id, i.in_reply_to_user_id)
    tweet_id.append(i.id)
```

```python
followers_id = []  
tmp = []
for response in tweepy.Paginator(client.get_users_followers, USER_ID, max_results=1000, limit=5):
    for i in response:
        tmp.append(i)
```

```python
following_id = []
tmp = []
for response in tweepy.Paginator(client.get_users_following, USER_ID, max_results=1000, limit=2):
    for i in response:
        tmp.append(i)
```

```python
for i in range(0, len(tmp), 4):
    for j in range(len(tmp[i])):
        following_id.append(tmp[i][j].id)
```

```python
reply_user = []
for k in range(len(tweet_id)):
    query = 'conversation_id:' + str(tweet_id[k])
    l = client.search_recent_tweets(query=query, max_results = 100, expansions = ['author_id'])
    try:
        for i in l.data:
            reply_user.append(i.author_id)
    except:
        print("", end = "")
```

```python
liking_users = []
for k in range(len(tweet_id)):
    l = client.get_liking_users(id = tweet_id[k])
    try:
        for i in l.data:
            liking_users.append(i.id)
    except:
        print("", end = "")
```

```python
retweeters = []
for k in range(len(tweet_id)):
    l = client.get_retweeters(id = tweet_id[k])
    try:
        for i in l.data:
            retweeters.append(i.id)
    except:
        print("", end = "")
```

```python
total = {}
for i in liking_users:
    if i in followers_id:
        if i in total.keys():
            total[i] += 1;
        else:
            total[i] = 1
        
for i in retweeters:
    if i in followers_id:
        if(i in total.keys()):
            total[i] += 1;
        else:
            total[i] = 1
    
for i in reply_user:
    if i in followers_id:
        if i in total.keys():
            total[i] += 3;
        else:
            total[i] = 3
for i in following_id:
    if i in followers_id:
        if i in total.keys():
            total[i] += 10;
        else:
            total[i] = 10
```
